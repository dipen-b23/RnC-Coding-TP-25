```python
for i in df.columns[3:]:
   Q1 = df[i].quantile(0.25)
   Q3 = df[i].quantile(0.75)
   IQR = Q3 - Q1
   lower_bound = Q1 - 1.5 * IQR
   upper_bound = Q3 + 1.5 * IQR
   outlier=(df[i]<lower_bound) | (df[i]>upper_bound)
   df.loc[outlier,i]=df[i].median()

df
```
