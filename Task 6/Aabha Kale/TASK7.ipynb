```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Cereals.csv')
df.head()
```

```python
cols=df.select_dtypes(include=['number'])
new=cols.describe().loc[['min', '25%', '50%', '75%', 'max']]
print(new)
```

```python
df.replace(-1, pd.NA, inplace=True)
for a in cols.columns:
  df[a] = df[a].fillna(df[a].mean())
```

```python
ncols=df.select_dtypes(include=['number'])
nnew=ncols.describe().loc[['min', '25%', '50%', '75%', 'max']]
print(nnew)
```

```python
import numpy as np

def treat_noise(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    median = series.median()
    return series.apply(lambda x: median if x < lower or x > upper else x)

for a in cols.columns:
    df[a] = treat_noise(df[a])
```
