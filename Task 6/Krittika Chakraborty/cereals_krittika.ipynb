```python
for i in range(4,11):
  df.iloc[ : ,i]=df.iloc[ : ,i].replace(-1,df.iloc[ : ,i].mean())
```

```python
df
```

```python
df.describe()
```

```python
for i in df.columns[3:]:
   Q1 = df[i].quantile(0.25)
   Q3 = df[i].quantile(0.75)
   IQR = Q3 - Q1
   lower_bound = Q1 - 1.5 * IQR
   upper_bound = Q3 + 1.5 * IQR
   outlier=(df[i]<lower_bound) | (df[i]>upper_bound)
   df.loc[outlier,i]=df[i].median()
```

```python
hot=df[df['type']=="H"]
cold=df[df['type']=="C"]
print("mean rating of hot cereals=",hot['rating'].mean())
print("mean rating of cold cereals=",cold['rating'].mean())
if(hot['rating'].mean()>cold['rating'].mean()):
  print("hot cereals are prefered")
else:
  print("cold cereals are prefered")
```

```python
import matplotlib.pyplot as plt
shelf_counts = df['shelf'].value_counts()
plt.bar(shelf_counts.index,shelf_counts,color="black")
plt.title('Number of Cereals by Shelf')
plt.xlabel('Shelf Number')
plt.ylabel('Number of Cereals')
plt.show()
```

```python
plt.plot(df.index,df['sugars'])
plt.title('Sugar Content across Different Brands')
plt.xlabel('Index')
plt.ylabel('Sugar content')
plt.show()
```

```python
print("Average rating of all cereals",df['rating'].mean())
print("Average rating of healthy cereals",df[(df['fiber']>3) & (df['sugars']<3)]['rating'].mean())
if(df[(df['fiber']>3) & (df['sugars']<3)]['rating'].mean()>df['rating'].mean()):
  print("healthy cereals are preferred")
else:
  print("healthy cereals are not preferred")
```
