```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Documents/Cereals.csv')
```

```python
dfirst = df.head(10)
```

```python
df.loc[:,'protein':'vitamins']<0
```

```python
cols = df.loc[:, 'protein':'vitamins'].select_dtypes(include='int64').columns
for i in cols:
    df.loc[df[i] < 0, i] = df[i].mean()
```

```python
df.head(10)
```

```python
print("summary")
df.head()
```

```python
col = df.loc[:, 'protein':'vitamins'].select_dtypes(include='number').columns
for i in col:
    s = df[i].dropna()
    Q1 = s.quantile(0.25)
    Q3 = s.quantile(0.75)
    IQR = Q3 - Q1
    if IQR == 0:
        continue
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    median_val = s.median()
    df.loc[(df[i] < lower_limit) | (df[i] > upper_limit), i] = median_val

print("SUMMARY: ")
df.head(5)
```

```python
type_counts = df['type'].value_counts()
plt.bar(type_counts.index, type_counts.values, color=['skyblue', 'pink'])
plt.title('Hot vs Cold Cereals')
plt.xlabel('Cereal Type')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
minimum = df['rating'].idxmin()
maximum = df['rating'].idxmax()
print("Best Cereal: ", df.loc[maximum, 'name'])
print("Worst Cereal: ", df.loc[minimum, 'name'])
```

```python
sns.kdeplot(df[df.type == 'H'].rating, fill = True, label = 'Hot', color = 'pink')
sns.kdeplot(df[df.type == 'C'].rating, fill = True, label = 'Cold',  color = 'skyblue')
plt.title("Hot vs Cold")
plt.legend()  
plt.show()

max = df.groupby('type')['rating'].mean()
types = {'H': 'Hot', 'C': 'Cold'}
print("People liked", types[max.idxmax()], "cereals more.")
```

```python
minimum = df['sugars'].idxmin()
maximum = df['fiber'].idxmax()
print("Highest Fibre: ", df.loc[maximum, 'name'])
print("Lowest Sugar: ", df.loc[minimum, 'name'])
```

```python
#How does the sugar content vary across different brands?
plt.figure(figsize=(15,5))
sns.barplot(x='name', y='sugars', data=df, palette = 'coolwarm')
plt.title("Sugar Content Across Cereal Brands")
plt.xlabel("Brand Name")
plt.ylabel("Sugar (grams per serving)")
plt.xticks(rotation=90)

plt.show()
```

```python
avgrating = df['rating'].mean()
high = df[df['rating'] > avgrating]['sugars'].mean()
low  = df[df['rating'] <= avgrating]['sugars'].mean()
print("Average sugar in high-rating cereals:", high)
print("Average sugar in low-rating cereals:",  low)
```

```python
fiber_median = df['fiber'].median()
sugar_median = df['sugars'].median()
healthy = df[(df['fiber'] > fiber_median) & (df['sugars'] < sugar_median)]
unhealthy = df[~((df['fiber'] > fiber_median) & (df['sugars'] < sugar_median))]
print("Mean rating of healthy cereals:", healthy['rating'].mean())
print("Median rating of unhealthy cereals:", unhealthy['rating'].median())
```
