```python
import pandas as pd
import numpy as np
```

# cleaning the data

```python

ins = pd.read_csv('insurance.csv')
```

```python
ins
```

```python
ins.describe()
```

```python
for col in ['age', 'bmi', 'children','region_Numeric','expenses']:
    ins[col] = (ins[col] - ins[col].min()) / (ins[col].max() - ins[col].min())
```

```python
ins['sex_Numeric'] = ins['sex'].map({'male': 1, 'female': 0})
ins['Smoker_Numeric'] = ins['smoker'].map({'yes': 1, 'no': 0})
ins['region_Numeric']= ins['region'].map({'southwest':0,'southeast':1,'northwest':2,'northeast':3})
```

# main code

```python
lr = 0.1 # learning rate
er_r = 0.01 # acceptable error
N = len(ins)
M = np.ones(7) # all the weights and calcs all at the same time
# the code line given below turn them into a matrix to do matrix multiplication
x = np.c_[np.ones(N),ins['age'],ins['bmi'],ins['children'],ins['sex_Numeric'],
          ins['Smoker_Numeric'],ins['region_Numeric']]
y= ins['expenses'].values
while True:
    prid = x.dot(M) # this gives the predicted value
    er = prid - y
    g= (2/N) * x.T.dot(er) #this shows the gradiant
    if np.all(np.abs(g) < er_r):
        break
    M = M - (lr*g)
print("the weights are ",M)
```

```python
user_age = input("enter age")
user_bmi = input("enter bmi")
user_child = input("enter no of children")
user_sex = input("enter male as 1 and female as 0")
user_smoke = input("enter yes as 1 and no as 0 for smoking")
user_region = input("southwest as 0 ,southeast as 1 ,northwest as 2 and northeast as 4")
sol = 1000*(
        -0.17833174 + 
       float(user_age) * 0.20517121 + 
       float(user_bmi) *  0.35153736 + 
       float(user_child) * 0.13594588 - 
       float(user_sex) * 0.00286601 + 
       float(user_smoke) * 0.38568847 + 
       float(user_region) * 0.05048835)
print("the insureance is", sol)
```
