```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df=pd.read_csv(r"C:\Users\91905\Downloads\heart.csv")
```

```python
x=df[["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal"]].to_numpy()
y=df["target"].to_numpy()
```

```python
def sigmoid(z):
    return 1.0/(1.0+np.exp(-z))
```

```python
def gradient(theta, x, y):
    m=y.size
    return (x.T @ (sigmoid(x@theta)-y))/m
```

```python
def gradient_descent(x, y, alpha, num_iters, tol=1e-7):
    x_b=np.c_[np.ones((x.shape[0],1)),x]
    theta=np.zeros(x_b.shape[1])

    for i in range(num_iters):
        grad=gradient(theta,x_b,y)
        theta-=alpha*grad

        if np.linalg.norm(grad)<tol:
            break
    return theta   
```

```python
def predict_prob(x, theta):
    x_b=np.c_[np.ones((x.shape[0],1)),x]
    return sigmoid(x_b@theta)
```

```python
def predict(x, theta, threshold=0.5):
    return (predict_prob(x, theta)>=threshold).astype(int)
```

```python
m=x.shape[0]
split=int(.8*m)
x_train=x[:split]
x_test=x[split:]
y_train=y[:split]
y_test=y[split:]

x_train=np.array(x_train, dtype=float)
x_test=np.array(x_test, dtype=float)
mu=x_train.mean(axis=0)
sigma=x_train.std(axis=0)
epsilon=1e-8
sigma=np.where(epsilon>sigma, epsilon, sigma)

x_train=(x_train-mu)/sigma
x_test=(x_test-mu)/sigma
```

```python
theta_hat=gradient_descent(x_train, y_train, 0.0001, 1000)
y_pred_train=predict(x_train, theta_hat)
y_pred_test=predict(x_test, theta_hat)
```

```python
def accuracy(y_true, y_pred):
    c=0
    for i in range(len(y_true)):
        if y_true[i]==y_pred[i]:
            c+=1

    return c/len(y_true)
```

```python
train_acc=accuracy(y_train, y_pred_train)
test_acc=accuracy(y_test, y_pred_test)
print("Training accuracy:",train_acc)
print("Testing accuracy:",test_acc)
```
