```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

# Cleaning and Setting up data

```python
bst = pd.read_csv('Boston.csv')
```

```python
bst
```

```python
train , test = train_test_split(bst,test_size=.2,random_state=1) 
#used to get training and testing data
```

```python
predict = ['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio'
           ,'black','lstat']
target = 'medv'
```

```python
x = train[predict].copy()
y = train[target].copy()
```

```python
x_mean = x.mean()
x_std = x.std()
```

```python
x = (x-x_mean)/x_std
```

```python
x['inter'] = 1
x = x[['inter'] + predict] #this is used to add intercept into the matrix
```

```python
alpha = 3 #instead of lambda we are using alpha but they are the same thing here
I = np.identity(x.shape[1]) #identity matrix
I[0][0] = 0 # this is a rule in ridge regression
```

# Ridge coding

```python
er = alpha * I
```

```python
b = np.linalg.inv(x.T @ x + er) @ x.T @ y
# @ is used as multiplication for matrix
""""
We get this from x * b = y (where they are all matrix)
(x^-1 * x)*b = x^-1 * y
b = (x^-1 * x)^-1 * x^-1 * y (this is the same as the equation given above)
"""
```

```python
x_test = test[predict]
x_test = (x_test - x_mean)/x_std
x_test['inter'] = 1
x_test = x_test[['inter'] + predict]
```
