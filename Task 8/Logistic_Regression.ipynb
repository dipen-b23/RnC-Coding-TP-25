```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('heart.csv')

df.head()
```

```python
X= np.array(df.drop(columns= ['target']))
y= np.array(df['target'])
```

```python
X_train= X[:250, :]
y_train= y[:250]
X_test=X[250:, :]
y_test= y[250:]
```

```python
avg= np.mean(X_train, axis=0)
dev= np.std(X_train, axis=0)
X_rescaled= (X_train- avg)/dev
print(X_train[0])
X_test_rescaled = (X_test - avg) / dev
X_rescaled[0]
```

```python
learning_rate= 0.00001
iterations=1000
```

```python
m,n= X_train.shape
w=np.zeros(n)
b=0
```

```python
def fit():
    for i in range(iterations):
        update_weights()

def update_weights():
    global w, b
    z= np.dot(X_rescaled, w) +b
    sigmoid= 1/(1+np.exp(-z))
    dw= (1/m) * np.dot(X_rescaled.T, (sigmoid- y_train))
    db= (1/m) *np.sum(sigmoid- y_train)
    w= w- learning_rate* dw
    b= b- learning_rate* db

def predict():
    z= np.dot(X_test_rescaled, w) +b
    y_pred= 1/(1+np.exp(-z))
    y_pred= np.where(y_pred>0.5, 1, 0)
    return (y_pred)
    
```

```python
fit()
results = predict()
print("Predictions for X_test:")
print(results)
```
